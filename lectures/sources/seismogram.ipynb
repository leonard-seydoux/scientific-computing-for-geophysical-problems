# Seismograms examples for the slides

```python
import matplotlib.pyplot as plt
import obspy
```

## Parameters

```python
plt.rcParams["font.sans-serif"] = "Ubuntu"
plt.rcParams["font.size"] = 18
plt.rcParams["figure.facecolor"] = (0, 0, 0, 0)
plt.rcParams["savefig.transparent"] = True
plt.rcParams["savefig.bbox"] = "tight"
plt.rcParams["savefig.pad_inches"] = 0.01
plt.rcParams["savefig.dpi"] = 200
```

## Example of a daylong seismogram

```python
FILEPATH = "https://examples.obspy.org/GR.BFO..LHZ.2012.108"
stream = obspy.read(FILEPATH)

plt.figure(figsize=(8, 0.8))
ax = plt.axes([0, 0, 1, 1])
ax.plot(stream[0].data, lw=0.5)
ax.set_xlim([0, stream[0].stats.npts])
ax.set_ylim([stream[0].data.min(), stream[0].data.max()])
ax.set_axis_off()
plt.savefig("../images/seismogram.png")
```

```python
FILEPATH = "https://examples.obspy.org/GR.BFO..LHZ.2012.108"
COLORS = "#1e81b0", "#e28743", "#eab676", "#76b5c5"
stream = obspy.read(FILEPATH)

fig = plt.figure(figsize=(6, 2))
fig = stream[0].plot(
    type="dayplot",
    interval=60,
    vertical_scaling_range=8e3,
    one_tick_per_line=True,
    color=COLORS,
    show_y_UTC_label=False,
    fig=fig,
    size=(1000, 800),
    handle=True,
    title=None,
    linewidth=0.7
)
fig.savefig("../images/seismogram_long.png")
```

```python
FILEPATH = "https://examples.obspy.org/GR.BFO..LHZ.2012.108"
COLORS = "#1e81b0", "#e28743", "#eab676", "#76b5c5"
stream = obspy.read(FILEPATH)

fig = plt.figure(figsize=(16, 9))
fig.set_facecolor("black")
fig = stream[0].plot(
    type="dayplot",
    interval=20,
    vertical_scaling_range=8e3,
    color=COLORS,
    show_y_UTC_label=False,
    fig=fig,
    handle=True,
    title=None,
    linewidth=1, 
)

ax = fig.get_children()[1]
ax.set_facecolor("black")
ax.set_axis_off()
fig.savefig("../images/seismogram_verylong.png", transparent=False)
```
