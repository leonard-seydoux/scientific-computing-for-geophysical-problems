# Seismograms examples for the slides

```python
# coding: utf-8

import matplotlib.pyplot as plt
import numpy as np

MATH_COLOR = 5 / 255, 103 / 255, 177 / 255

plt.rcParams["figure.dpi"] = 300
plt.rcParams["mathtext.fontset"] = "dejavusans"

fig, ax = plt.subplots(1, figsize=(4, 3))

# Continuous case
t = np.linspace(0.6, 3 * np.pi, 200)
y = 0.9 * np.sin(2 * np.pi * 0.2 * t)

ax.plot(t, y, clip_on=False, lw=0.7, c="k")

for item in zip(t[::20], np.round(y[::20], 1)):
    ax.plot(*item, ".", clip_on=False, lw=0.7, c=MATH_COLOR)
    ax.plot(2 * [item[0]], [0, item[1]], lw=0.7, c=MATH_COLOR)


ax.axvline(t[40], ls="--", lw=0.6, color=MATH_COLOR)
ax.axvline(t[60], ls="--", lw=0.6, color=MATH_COLOR)
arrowprops = dict(arrowstyle="<->", clip_on=False, color=MATH_COLOR)
kwargs = dict(
    xycoords="data",
    textcoords="data",
    arrowprops=arrowprops,
)
ax.annotate("", (t[40], 0.5), xytext=(t[60], 0.5), **kwargs)
ax.text(
    t[50],
    0.55,
    r"$\Delta t$",
    usetex=True,
    ha="center",
    size=14,
    color=MATH_COLOR,
)


ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.set_xticks([])
ax.set_yticks([])
arrowprops = dict(arrowstyle="->", clip_on=False)
kwargs = dict(
    xycoords="data",
    textcoords="data",
    arrowprops=arrowprops,
)
ax.annotate("", (t.max(), 0), xytext=(0, 0), **kwargs)
ax.annotate("", (0.3, 1), xytext=(0.3, -1), **kwargs)

ax.text(-0.1, -0.13, r"$0$", usetex=True, size=14, color=MATH_COLOR)
ax.text(t.max(), -0.13, r"$t$", usetex=True, size=14, color=MATH_COLOR)
ax.text(0, 1.05, r"$u$", usetex=True, size=14, color=MATH_COLOR)

ax.set_ylim(-1, 1)
ax.set_xlim(0, t.max())

plt.savefig("../images/seismograms/digital.svg", transparent=True)
```
