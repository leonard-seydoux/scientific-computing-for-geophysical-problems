# Seismograms examples for the slides

```python
import matplotlib.pyplot as plt
import schemdraw

from schemdraw import flow
```

## Parameters

```python
plt.rcParams["savefig.dpi"] = 300
plt.rcParams["figure.dpi"] = 300
plt.rcParams["font.sans-serif"] = "Ubuntu"

schemdraw.config(fontsize=18)
with schemdraw.Drawing(file="../images/inverse.svg") as diagram:
    diagram += (a := flow.Start().fill("white").label("Data")) 
    diagram += (c := flow.Start().fill("white").label("Model").at((5, 0)))
    diagram += flow.Arc2(arrow='->').at(a.N).to(c.N).label("Inverse")
    diagram += flow.Arc2(arrow='<-', k=-0.5).at(a.S).to(c.S).label("Forward")
```

```python
plt.rcParams["savefig.dpi"] = 300
plt.rcParams["figure.dpi"] = 300
plt.rcParams["font.sans-serif"] = "Ubuntu"

schemdraw.config(fontsize=18)
with schemdraw.Drawing(file="../images/inversion.svg") as diagram:
    diagram += (a := flow.Box().fill("white").label("Model")) 
    diagram += (d := flow.Box().fill("white").label("Data").at((6, -4)))
    diagram += (c := flow.Box().fill("white").label("Synthetics").at((6, 0)))
    diagram += flow.Arrow(arrow='->').at(a.E).to(c.W).label("Forward")
    diagram += flow.Wire("c", k=(a.S[0] - 6)).at((6, -2)).to(a.S).label("Update")
    diagram += flow.Wire(arrow="->").at((a.S[0], a.S[1]-1)).to(a.S)
    diagram += flow.Wire(arrow='<->').at(c.S).to(d.N).label("Error")
    
```
