# Seismograms examples for the slides

```python
import matplotlib.pyplot as plt
import schemdraw

from schemdraw import flow
```

## Parameters

```python
plt.rcParams["savefig.dpi"] = 300
plt.rcParams["figure.dpi"] = 300
plt.rcParams["font.sans-serif"] = "Cascadia Code"

schemdraw.config(fontsize=18)
with schemdraw.Drawing(file="../images/flowcharts/seismograph_circuit.svg") as diagram:
    diagram += flow.Arrow().length(1)
    diagram += (a := flow.Box().fill("white").label("Analog\nto digital\nconverter")) 
    diagram += flow.Arrow().at(a.E).length(1)
    diagram += (c := flow.Box().fill("white").label("Computer"))
    diagram += flow.Arrow().at(c.E).length(1)
    diagram += (d := flow.Box().fill("white").label("Data\ncenter"))
    diagram += (g := flow.Box().fill("white").label("GPS").at((3, -3)))
    diagram += (p := flow.Box().fill("white").label("Power\nsupply").at((7, -3)))
    diagram += flow.Arrow().length(1).at(p.N).to(c.S)
    diagram += flow.Arrow().length(1).at(g.N).to(c.S)
```
