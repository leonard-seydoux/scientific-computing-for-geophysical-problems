```python
# coding: utf-8

import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = "Ubuntu"
plt.rcParams["font.size"] = 18
plt.figure(figsize=(4, 5))

labels = " ", "Notebooks", "Final Exam (2h)"
sizes = [50, 25, 25]
colors = ["lightsteelblue", "gold", "orange"]

plt.pie(
    (50, 50),
    labels=("AF", "LS"),
    colors=colors,
    radius=1.2, 
)

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct="%1d%%",
    wedgeprops=dict(linewidth=1, edgecolor="k"),
)

# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis("equal")
plt.savefig("../images/grading.svg", transparent=True, bbox_inches="tight")
```
